```python
for ID, packet_def in packets.items():
    packet_name = ""
    if packet_def.repeat > 1:
        packet_name = f"{packet_def.name[ID-packet_def.ids[0]]}_{packet_def.offset*(ID-packet_def.ids[0])}"
    else:
        packet_name = f"{packet_def.name[0]}"
    print(packet_name)
    header_str = ""
    for field in packet_def.data:
        if field['type'] != 'bitfield':
            header_str += field['name'] + ", "
        else:
            for bit in field['bits']:
                header_str += field['name'] + ":" + bit['name'] + ", "
        
    print(header_str)
```
